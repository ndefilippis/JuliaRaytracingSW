```julia
using Interpolations
using CairoMakie
```

```julia
N = 3
f(x) = x^(-N)

xs = 0.1:0.05:100.0
f_vals = f.(xs)

dx = 0.1
eval_points = 1:dx:10
```

```julia
scalings = exp.(log(0.1):0.1:log(10.0))
Nscales = length(scalings)
values1 = zeros(Nscales)
values2 = zeros(Nscales)
for idx=1:Nscales
    a=scalings[idx]
    f_itp = linear_interpolation(a * xs, f_vals)
    values1[idx] = sum(f_itp.(eval_points) * dx)
    f_new(x) = f(x / a)
    values2[idx] = sum(f_new.(eval_points) * dx)
end
```

```julia
fig = Figure()
ax = Axis(fig[1,1]; xscale=log10, yscale=log10)
lines!(ax, scalings, values1)
lines!(ax, scalings, values2)
lines!(ax, scalings, scalings.^N; linestyle=:dash, color=:black)

fig
```
