```julia
using AbstractFFTs
using CairoMakie
```

```julia
k₀ = 1.5
ω₀ = 0.96
x = 4
dt = 2e-1
t = 0:dt:400
N = length(t)

samples = @. exp(1im * (k₀*x - ω₀*t))
ω = fftshift(fftfreq(N, 2π/dt))
result = fftshift(fft(samples))
fig = Figure()
ax = Axis(fig[1,1])
lines!(ax, ω, abs2.(result))
lines!([ω₀, ω₀], [0, 3e6])
lines!([-ω₀, -ω₀], [0, 3e6])
fig
```
