```julia
using FourierFlows
using CairoMakie
using AbstractFFTs
```

```julia
grid = OneDGrid(CPU(), Lx=2π, nx=256)
k0 = 10
k1 = 14
U = zeros(Float64, grid.nx)
for k=-k1:k1
    if k0^2 <= k^2 && k^2 <= k1^2
        phase = rand()
        amp = randn()
        @. U += amp * cos(k*grid.x + 2*π*phase)
    end
end
```

```julia
plot(grid.x, abs2.(U))
```

```julia
Uh = rfft(U)
plot(grid.kr[:], log10.(abs2.(Uh)))
```
