```julia
using CairoMakie
```

```julia
function P(q)
    return (0.01 < q) * (q < 100) * q^(-5)
end

function C(q, σ)
    σ₀ = 1
    σd = 100
    σ = abs(σ)
    return ((σ < σ₀) * σ₀^(-2) + (σ > σ₀) * (σ < σd) * σ^(-2)) * P(q)
end
```

```julia
q = 1.0:0.1:300
dq = 0.1
```

```julia
fig = Figure()
ax = Axis(fig[1,1]; xscale=log10, yscale=log10)
lines!(ax, q, replace(C.(q, q), 0.0=>NaN))

fig
```

```julia
c = 3

function compute_integral(x)
    dη = 0.002
    η = (0:dη:2π)'
    
    dq = 0.5
    q = 1.0:dq:(min(100, x / c * 100)*3)

    return c * x / (2 * (2π)^2) * sum(@. q^5 * cos(η)^2 * sin(η)^2 * C(q, -c * q * cos(η) / x) * dq * dη)
    
end
```

```julia
x = exp.(log(0.02):0.1:log(300.0))
bs = compute_integral.(x)
```

```julia
fig = Figure()

ax = Axis(fig[1,1]; xscale=log10, yscale=log10)
lines!(ax, x, bs)
lines!(ax, x, bs[10] * (x/x[10]).^3, linestyle=:dash, color=:black)
lines!(ax, x, bs[end] * (x/x[end]).^1, linestyle=:dash, color=:black)


fig
```
