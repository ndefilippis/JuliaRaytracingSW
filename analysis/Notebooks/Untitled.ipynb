```julia
using CairoMakie
using FourierFlows
using AbstractFFTs
```

```julia
grid = TwoDGrid(; nx=512, Lx=2π)
K = fftshift(sqrt.(grid.Krsq[15, :]))
Nω = 3000
ω_max = 100
ω = ω_max * ((1:Nω)/Nω)
values = zeros(grid.nkr, Nω)
k_idx = @. floor(Int, grid.nkr * K / grid.kr[end]) + 1
for i=1:grid.nkr
    if k_idx[i] <= grid.nkr
        values[k_idx[i], :] .+= 1
    end
end
heatmap(grid.kr[:], ω, values)
```
